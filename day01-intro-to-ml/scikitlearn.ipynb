```python
from sklearn.datasets import  load_breast_cancer
from sklearn.model_selection import train_test_split #function to split data into training portion
from sklearn.preprocessing import StandardScaler #function for scaling data
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = load_breast_cancer(return_X_y=True)
```

Loading Data

```python
X, y= load_breast_cancer(return_X_y=True)
```

```python
y
```

Split data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

Scaling the data

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Training the data

```python
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
```

Prediction

```python
print(knn.score(X_test_scaled,y_test))
```

## **SPLITTING DATA**

### Why Splitting Data is Necessary

Imagine you're studying for a test. If you only practiced with the exact same questions that will be on the test, you might do well, but would you truly understand the material and be able to answer new questions?

Splitting our data is similar. We divide our dataset into two main parts:

1.  **Training Data:** This is the data we use to "teach" our machine learning model. The model learns patterns and relationships from this data.
2.  **Testing Data:** This is data that the model has *never* seen before. We use this data to evaluate how well our trained model performs on new, unseen examples.

By testing on separate data, we get a more realistic idea of how our model will perform in the real world, where it will encounter new data. If we trained and tested on the same data, we might overestimate how good our model is because it has simply memorized the answers.

So, splitting the data helps us ensure our model can generalize and make accurate predictions on new data, not just the data it was trained on.

```python
# Common imports
import numpy as np
from sklearn.model_selection import (
    train_test_split, KFold, RepeatedKFold, StratifiedKFold, RepeatedStratifiedKFold,
    ShuffleSplit, StratifiedShuffleSplit,
    GroupKFold, GroupShuffleSplit, StratifiedGroupKFold,
    LeaveOneOut, LeavePOut, LeaveOneGroupOut, LeavePGroupsOut,
    PredefinedSplit, TimeSeriesSplit, check_cv
)

# Sample dataset
X = np.arange(40).reshape(20, 2)   # 20 samples, 2 features
y = np.array([0, 1, 0, 1, 0, 1])  # binary target
groups = [1, 1, 2, 2, 3, 3]       # group labels
```

🔹 1. Train-Test Split

The simplest way to split data into train and test sets.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Train indices:", X_train)
print("Test indices:", X_test)
```

🔹 2. KFold

Splits into k folds, cycling each as test set.

```python
kf = KFold(n_splits=5, shuffle=False)
for fold, (train, test) in enumerate(kf.split(X), 1):
    print(f"Fold {fold} → Train: {train}, Test: {test}")
```

🔹 3. RepeatedKFold

Repeats K-Fold multiple times with different splits.

```python
rkf = RepeatedKFold(n_splits=3, n_repeats=2, random_state=42)
for train, test in rkf.split(X):
    print("Train:", train, "Test:", test)
```

🔹 4. StratifiedKFold

Keeps class distribution balanced across folds.

```python
skf = StratifiedKFold(n_splits=3)
for train, test in skf.split(X, y):
    print("Train:", train, "Test:", test)
```

🔹 6. ShuffleSplit

Random train/test splits multiple times.

```python
ss = ShuffleSplit(n_splits=3, test_size=0.3, random_state=42)
for train, test in ss.split(X):
    print("Train:", train, "Test:", test)
```

🔹 7. StratifiedShuffleSplit

Random splits but preserves class balance.

```python
sss = StratifiedShuffleSplit(n_splits=3, test_size=0.3, random_state=42)
for train, test in sss.split(X, y):
    print("Train:", train, "Test:", test)
```
